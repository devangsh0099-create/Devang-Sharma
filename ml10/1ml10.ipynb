```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='green')
```

```python
y_pred = ((78.35 * X) + 100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='#00a65a',label='OLS')
plt.plot(X,y_pred,color='red',label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 100
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='#00a65a',label='OLS')
plt.plot(X,y_pred1,color='red',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = (78.35 * X + b).reshape(-1)

plt.figure(figsize=(10,10))


plt.scatter(X, y)


plt.plot(X, reg.predict(X), color='red', label='OLS')

plt.plot(X, y_pred3, label=f'b = {b}')
plt.plot(X, y_pred1, label='b = b1')
plt.plot(X, y_pred, label='b = 0')

plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01
epochs = 100
for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)
  y_pred = m * X + b
  plt.plot(X,y_pred)
plt.scatter(X,y)
```
